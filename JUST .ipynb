### Stop Words

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
c=0
for word in doc1:
    if word.is_stop==True:
        print(word)
        c=c+1
print(c)
```

```python
list1=[]
for word in doc1:
    if word.is_stop==False and word.is_punct==False:
        print(word)
        list1.append(word)
```

```python
list1
```

```python
new_text=""
for word in doc1:
    if word not in STOP_WORDS:
        new_text=new_text+" "+word
```

```python
#creating a dataframe
cols=['Token','POS','Explain_POS','Tag','Explain_Tag']
cols
rows=[]
for token in doc1:
    row=token,token.pos_,spacy.explain(token.pos_),token.tag_,spacy.explain(token.tag_)
    rows.append(row)
rows
```

```python
import pandas as pd
token_df=pd.DataFrame(rows,columns=cols)
token_df
```

```python
token_df['POS'].value_counts()
```

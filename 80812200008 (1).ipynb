```python
# Creating a doc object

doc=nlp(text)
```

```python
doc
```

```python
# Number of sentences

sent_count=0
for sent in doc.sents:
    sent_count=sent_count+1
    print(sent_count,'==>',sent)
print('The total number of sentences is:',sent_count)
```

```python
# Tokenization

c=0
for token in doc:
    print(token)
    c+=1
print('The number of tokens are:',c)
```

```python
# Parts of speech

for token in doc:
    print(token,'==>',token.pos_)
```

```python
# tokens and POS where token is not a stop word, punctuation and a number

for token in doc:
    if token.is_stop==False and token.is_punct==False and token.is_digit==False:
        print(token,'==>',token.pos_)
```

```python
# Count of number of nouns and verbs

pos_count=doc.count_by(spacy.attrs.POS)
for x,y in sorted(pos_count.items()):
    print(x,doc.vocab[x].text,y)
```

Number of verbs = 136

### Question 2

```python
# Named entities and their label

for ent in doc.ents:
    print(ent.text,'==>',ent.label_)
```

```python
# Visualising the labels

from spacy import displacy
displacy.render(doc, style='ent')
```

```python
# Creating a DataFrame of entities, types , explanation of types

import pandas as pd
rows=[]
cols=['Name_of_Entity','Entity_type','Explation_of_type']
for ent in doc.ents:
    row=ent,ent.label_,spacy.explain(ent.label_)
    rows.append(row)
ent_df=pd.DataFrame(rows,columns=cols)
ent_df
```

```python
# Creating a plot to show the count of each type

import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,12))
ax = sns.countplot(x=ent_df['Entity_type'])
for p in ax.patches:
   ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.show();
```

### Entity appearing most number of times:
    
From the above count plot it can be seen that, Person entity is the most occuring one with 53 times

```python
# Grouping based on Persons

ent_filter=ent_df[ent_df['Entity_type']=='PERSON']
ent_filter
```

```python
# counting each names in the filtered data

from collections import Counter
most_ent=[]
for ent in doc.ents:
    most_ent.append(ent.text)
Counter(most_ent).most_common(10)
```

### Inference 

From the ent_filter tabel and the top 10 most common list, it can be seen that 'Jim' is a character and it is occuring the most number of times.

### Question 3

```python
# Checking the number of times the phrase 'Article 370' has occured

from spacy.matcher import Matcher

matcher_1=Matcher(nlp.vocab)
pattern_1=[{'text':'Article'},{'text':'370'}]
matcher_1.add('Pattern1',[pattern_1])
match_1=matcher_1(doc)
print('''The number of times the phrase 'Article 370' has occured:''',len(match_1))
```

```python
# Occurence of ‘depict’, ‘reveal’, ‘discover’, ‘direct’, ‘narrate’, ‘illustrate’ and ‘portray’

matcher_2=Matcher(nlp.vocab)
pattern_2=[{'LEMMA':
            {'IN':['depict','reveal','discover','direct','narrate','illustrate','portray']}}]

matcher_2.add('Pattern2',[pattern_2])
match_2=matcher_2(doc)
```

```python
for match_id,start,end in match_2:
    span=doc[start:end]
    print(span)
```

### Inference

The word that is most used to establish ideas is 'Reveal'.

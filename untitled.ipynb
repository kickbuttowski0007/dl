```python
# Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the csv file

credit=pd.read_csv('Credit_Card_L1Y5s13EN8.csv')
credit
```

```python
credit.info()
```

## Preprocessing

```python
# Dropping rows with null values

credit=credit.dropna()
credit
```

```python
# Dropping the column CUST_ID

credit=credit.drop(['CUST_ID'],axis=1)
credit
```

```python
# Dropping rows with tenure < 10

credit1=credit[credit['TENURE']>=10]
credit1
```

```python
credit1['TENURE'].min(),credit1['TENURE'].max()
```

```python
# Choosing Tenure as target

y=credit1['TENURE']
y
```

```python
x=credit1.drop(['TENURE'],axis=1)
x
```

```python
# Normalizing the features

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_scaled=sc.fit_transform(x)
x_scaled
```

```python
x_scaled_df=pd.DataFrame(x_scaled,columns=x.columns)
x_scaled_df
```

```python
# Dimension of the feature set

x_scaled_df.shape
```

## Principle Component Analysis

```python
# Construction of covariance matrix

cm=np.cov(x_scaled.T)
cm
```

```python
cm.shape
```

```python
# Finding the eigen values

eig_val,eig_vec=np.linalg.eig(cm)
eig_val
```

```python
# Finding the eigen vector

eig_vec
```

```python
# Sorting and finding the explained variance ratio of each component

## Sorting
sorted_eig_val=[i for i in sorted(eig_val, reverse=True)]

## Calculating the total
tot=sum(sorted_eig_val)

## Calculating the ratio explained
exp_var=[(i/tot) for i in sorted_eig_val]
exp_var
```

```python
# Construction of projection matrix and choosing only 2 dimensions

## Projection matrix
eigen_pair=[(np.abs(eig_val[i]),eig_vec[:,i]) for i in range(len(eig_val))]

## Taking 2 dimensions
w=np.hstack((eigen_pair[0][1][:,np.newaxis],eigen_pair[1][1][:,np.newaxis]))
w
```

```python
w.shape
```

```python
# Transforming 16 dimensions to 2 dimensions

x_scaled.shape , w.shape
```

```python
# Dot product

pca_x=x_scaled.dot(w)
pca_x
```

```python
pca_x.shape
```

```python
# Visualising the ttransformed dataset

plt.figure(figsize=(12,12))
plt.scatter(pca_x[:,0],pca_x[:,1],c=y)
plt.show();
```

### Comments:

It can be seen that after feature reduction to 2 dimensions, there is no proper segregation between different tenure years.

This could be because, the first 2 components explain only 50% of the total variance.
